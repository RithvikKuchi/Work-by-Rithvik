```python
#creating the data series - 1d arry (rows)
import pandas as pd
stu_dict={"stu_reg":[1,2,3,4,5],
          "stu_name":["ravi","raghu","ramya","sai","anu"],
          "stu_res":["pass","fail","pass","fail","with held"]}
stu_s=pd.Series(stu_dict)
print(stu_s)
print("***************")
print(stu_s["stu_name"])
print("$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$")
for ele in stu_s:
  print(ele)
```
