**MATPLOATLIB** **BASICS**

````python
#Simple line plot
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

plt.plot(x, y)
plt.show()
````

````python
#Line Plot with Marker and Style
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [5, 15, 25, 35, 45]

plt.plot(x, y, marker='o', linestyle='--')
plt.show()
````

````python
#Add Title and Labels
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 40, 20, 30]

plt.plot(x, y)
plt.title("Sample Graph")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
````

````python
#Bar Graph
import matplotlib.pyplot as plt

x = ["Mon", "Tue", "Wed", "Thu"]
y = [10, 20, 15, 25]

plt.bar(x, y)
plt.show()
````

````python
#Pie Chart
import matplotlib.pyplot as plt

labels = ["A", "B", "C", "D"]
sizes = [25, 35, 20, 20]

plt.pie(sizes, labels=labels)
plt.show()
````

````python
#Scatter Plot
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [5, 20, 14, 32, 22]

plt.scatter(x, y)
plt.show()
````

````python
#Histogram
import matplotlib.pyplot as plt

data = [10, 20, 20, 30, 40, 40, 40, 50]

plt.hist(data)
plt.show()
````

````python
#Multiple Lines in One Graph
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [40, 30, 20, 10]

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
````

````python
#plotting for a single value
import matplotlib.pyplot as plt
x=[10,56,7]
y=[20,78,89]
plt.plot(x,y,marker="^",ms=10)
plt.draw()
````

````python
#plotting for a single value
import matplotlib.pyplot as plt
x=[10,56,7]
y=[20,78,89]
plt.plot(x,y,ms=10)
plt.draw()
````

````python
#plotting multiple values

import matplotlib.pyplot as plt

x = ["mon", "tue", "wed", "thu", "fri", "sat", "sun"]
y = [33, 22, 11, 44, 55, 67, 8]   # Fixed length (7 values)

plt.plot(x, y, "h-.k", mec="y", mfc="y")   # h = hexagon marker
plt.show()
````
