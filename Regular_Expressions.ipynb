**Regular Expression**

1.Find numbers in text

2.Check email format

3.Search a word in sentence

4.re means regular expression

5.It is used to find or check patterns in text

```python
#Basic
import re

text = "My number is 123"
print(re.findall("\d+", text))
```

```python
import re
s="welcome to python programming"
print(re.findall("p",s))
print(s)
```

```python
#Match A Pattern
import re

text = "hello world"
result = re.match("hello", text)

if result:
    print("Match found")
```

```python
#Search a pattern
import re

text = "I love python"
result = re.search("python", text)

if result:
    print("Found")
```

```python
#Find all Matches
import re

text = "cat bat mat cat"
result = re.findall("cat", text)

print(result)
```

```python
#Replace Text
import re

text = "I like Java"
new_text = re.sub("Java", "Python", text)

print(new_text)
```

```python
#Split String Using Pattern
import re

text = "apple,banana;grapes orange"
result = re.split("[,; ]", text)

print(result)
```

```python
#Check Digits
import re

text = "My number is 98765"
result = re.findall("\d", text)

print(result)
```

```python
#Check Email Pattern (Basic)
import re

email = "test@gmail.com"
pattern = r"\S+@\S+\.\S+"

if re.match(pattern, email):
    print("Valid Email")
```

```python
#Check Only Alphabets
import re

text = "HelloWorld"

if re.match("^[A-Za-z]+$", text):
    print("Only Alphabets")
```

```python
#Extract Numbers from Text
import re

text = "Order number is 456 and price is 789"
numbers = re.findall("\d+", text)

print(numbers)
```
