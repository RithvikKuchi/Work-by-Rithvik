# Files Notes In Python
1.Used to store data permanently.

2.Use open("filename", "mode") to access a file.

3.Common modes: 'r' (read), 'w' (write), 'a' (append), 'x' (create).

4.Use with open(...) for safe file handling.

5.read(), readline(), readlines() to read files.

6.write(), writelines() to write files.

7.Always close the file (or use with).

8.Text files = default, Binary files = use 'b' mode.

9.Errors: FileNotFoundError, PermissionError are common.

10.Combine modes like 'rb', 'w+', 'a+' for more control.

**TO PRINT FIRST 50 N EVEN NUMBERS**

```python
f = open("even.txt", "w")
for i in range(0,50):
    if i % 2 == 0:
        f.write(str(i) + "\n")
f.close()
f = open("even.txt", "r")
print(f.read())
f.close()
```

**TO PRINT FIRST 50 N ODD NUMBERS**

```python
f = open("odd.txt", "w")
for i in range(0,50):
    if i % 2 != 0:
        f.write(str(i) + "\n")
f.close()
f = open("odd.txt", "r")
print(f.read())
f.close()
```
